```python
import types
import pandas as pd
from botocore.client import Config

import ibm_boto3
import matplotlib
def __iter__(self): return 0

# @hidden_cell
# The following code accesses a file in your IBM Cloud Object Storage. It includes your credentials.
# You might want to remove those credentials before you share the notebook.
client_de52382743884d4bba886c929d00172d = ibm_boto3.client(service_name='s3',
    ibm_api_key_id='buKGiNWRapnKVt5Lrn00GX-xwXdZX7wONaAWCs8d00Vs',
    ibm_auth_endpoint="https://iam.ng.bluemix.net/oidc/token",
    config=Config(signature_version='oauth'),
    endpoint_url='https://s3-api.us-geo.objectstorage.service.networklayer.com')

body = client_de52382743884d4bba886c929d00172d.get_object(Bucket='reactivamadridapp-donotdelete-pr-rzg7nplfhrfogi',Key='Codvid19_Madrid.csv')['Body']
# add missing __iter__ method, so pandas accepts body as file-like object
if not hasattr(body, "__iter__"): body.__iter__ = types.MethodType( __iter__, body )
df_data_2 = pd.read_csv(body,sep=";",index_col='fecha_informe')
df_data_2.head()
```

```python
data_2=df_data_2[['municipio_distrito','casos_confirmados_totales','casos_confirmados_ultimos_14dias']]
data_2.dropna(inplace=True)
##cambiar los municipios por una variable categórica
data_2['municipio_distrito_discreto']=pd.factorize(data_2.municipio_distrito)[0]
data_2.head()
```

```python
from sklearn.preprocessing import StandardScaler
X=data_2.values[:,1:]
Clus_dataSet=StandardScaler().fit_transform(X)
Clus_dataSet
```

```python
from sklearn.cluster import KMeans
clusterNum=6
k_means=KMeans(init="k-means++", n_clusters=clusterNum,n_init=12)
k_means.fit(X)
labels=k_means.labels_
print(labels)
```

```python
data_2["cluster"]=labels
data_2.head(5)
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

fig = plt.figure(1, figsize=(12, 6))
plt.scatter(X[:,2],X[:,0], c=labels.astype(np.float), alpha=0.5)
plt.xlabel("municipio")
plt.ylabel("casos totales")
```

```python
Clus_dataSet.shape[0]
```

```python
import scipy
leng=Clus_dataSet.shape[0]
D=scipy.zeros([leng,leng])
for i in range (leng):
    for j in range(leng):
        D[i,j]=scipy.spatial.distance.euclidean(Clus_dataSet[i],Clus_dataSet[j])
        
```

```python
import pylab
import scipy.cluster.hierarchy
Z=scipy.cluster.hierarchy.linkage(D,'complete')
```

```python
from scipy.cluster.hierarchy import fcluster
max_d=10
clusters=fcluster(Z,max_d,criterion='distance')
clusters
```

```python
fig=pylab.figure(figsize=(12,20))

dendro=scipy.cluster.hierarchy.dendrogram(Z, leaf_rotation=0,leaf_font_size=12, orientation='right')
```
